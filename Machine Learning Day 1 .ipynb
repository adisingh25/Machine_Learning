```python
import matplotlib.pyplot as plt
import numpy
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.scatter(x,y)

plt.show()
```

```python
# here 5.0 is the mean , 1.0 is the standard deviation and 1000 values will be flooded in the x list
x = numpy.random.normal(5.0,1.0,1000)
y = numpy.random.normal(10.0,2.0,1000)

plt.scatter(x,y)
plt.show()
```

```python
x = numpy.random.normal(5.0,1.0,100)
y = numpy.random.normal(10.0,2.0,100)

plt.scatter(x,y)
plt.show()
```

```python
#selecting the training data 
# for training we are picking values from 0 to 79 
train_x = x[:80]
train_y = y[:80]

# for testing we are picking values from 80 to last 
test_x = x[80:]
test_y = x[80:]
```

```python
#plotting the training data
plt.scatter(train_x, train_y)
plt.show()
```

```python
plt.scatter(test_x,test_y)
plt.show()
```

```python
mymodel = numpy.poly1d(numpy.polyfit(train_x, train_y,4))
myline = numpy.linspace(0,6,100)

plt.scatter(train_x, train_y)
plt.plot(myline, mymodel(myline))
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(train_y, mymodel(train_x))
print(r2)

#r2 means R-squared. It measures the relationship between the x-axis and the y-axis.
# values usually ranges from 0 to 1 ....0 means no relation and 1 means complete (totally) related
```

```python
#testing the model with test data
r2 = r2_score(test_y, mymodel(test_x))

print(r2)
```

```python
# predicting the values on the basis of model created 
print(mymodel(5))
```

```python
# Taking about numpy.poly1d 
p = numpy.poly1d([1,2,3])
print(p)   # x^2 + 2x + 3
print(p(5))
```

```python
# Print the roots 
print(p.r)
```

```python
# print the coefficients 
print(p.c)
```

```python
print(p.order) # order of polynomial
```

```python
#construct ploynomial from the roots 
p = numpy.poly1d([1,2], True)
print(p)
# here 1 and 2 are the roots of the ploynomial whereas earlier we were sending the co-efficient
```

```python
#CREATING CONFUSION MATRIX 
```

```python
actual = numpy.random.binomial(1,0.9,size=100)
predicted = numpy.random.binomial(1,0.9,size=100)
```

```python
from sklearn import metrics
confusion_matrix = metrics.confusion_matrix(actual,predicted)
```

```python
print(confusion_matrix)
```

```python
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels = [False, True])
```

```python
cm_display.plot()
plt.show()
```

```python
accuracy = metrics.accuracy_score(actual,predicted)
print(accuracy)
```

```python
precision = metrics.precision_score(actual,predicted)
print(precision)
```
